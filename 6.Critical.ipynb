```python
x, y = meshgrid(arange(-1.5,1.5,0.1), arange(-1.5,1.5,0.1))
```

```python
xdot = -y -x**3
ydot = x - y**3
```

```python
quiver(xdot, ydot)
```

```python
xdot = -y + x**3
ydot = x + y**3
```

```python
quiver(xdot, ydot)
```
